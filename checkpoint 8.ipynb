```python
import pandas as pd
df = pd.read_csv("CC GENERAL.csv")
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df=df.drop("CUST_ID",axis=1)
df
```

```python
from sklearn.cluster import AgglomerativeClustering #Importing our clustering algorithm : Agglomerative
model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(df)
```

```python
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

```python
import matplotlib.pyplot as plt
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (df ['CREDIT_LIMIT'] , df ["PURCHASES"] , c= agglomerative[0], s=50)
ax.set_title("Agglomerative Clutering")
ax.set_xlabel("CREDIT_LIMIT")
ax.set_ylabel("PURCHASES")
plt.colorbar(scatter)
```

```python
df2=df.loc[df ["PURCHASES"]<20000]
```

```python
df2["PURCHASES"].value_counts()
```

```python
from sklearn.cluster import AgglomerativeClustering #Importing our clustering algorithm : Agglomerative
model=AgglomerativeClustering(n_clusters=10, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(df2)
```

```python
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

```python
import matplotlib.pyplot as plt
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (df2 ['CREDIT_LIMIT'] , df2["PURCHASES"] , c= agglomerative[0], s=10)
ax.set_title("Agglomerative Clutering")
ax.set_xlabel("CREDIT_LIMIT")
ax.set_ylabel("PURCHASES")
plt.colorbar(scatter)
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Customer Dendrograms")
dend=shc.dendrogram(shc.linkage(df2, method="complete"))
```

```python
from sklearn.cluster import KMeans  #Importing our clustering algorithm: KMeans
kmeans=KMeans(n_clusters=5, random_state=0)  #Cluster our data by choosing 5 as number of clusters
kmeans.fit(df)
```

```python
labels=pd.DataFrame(kmeans.labels_)
labels
```

```python
kmeans.predict(df)
print(kmeans.cluster_centers_)   #Printing the coordinates of cluster centers.
```

```python
kmeans.labels_
```

```python
df2['label']=kmeans.predict(df2)
```

```python
df
```

```python
plt.scatter(df2 ['CREDIT_LIMIT'][df2.label == 0],          
            df2["PURCHASES"][df2.label == 0],s=10,c='magenta')
plt.scatter(df2 ['CREDIT_LIMIT'][df2.label == 1],
           df2["PURCHASES"][df2.label == 1],s=10,c='yellow')
plt.scatter(df2 ['CREDIT_LIMIT'][df2.label == 2],
           df2["PURCHASES"][df2.label == 2],s=10,c='green')
plt.scatter(df2 ['CREDIT_LIMIT'][df2.label == 3], 
           df2["PURCHASES"][df2.label == 3],s=10,c='cyan')
plt.scatter(df2 ['CREDIT_LIMIT'][df2.label == 4], 
           df2["PURCHASES"][df2.label == 4],s=10,c='burlywood')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='red', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('CREDIT_LIMIT')
plt.ylabel('PURCHASES')
plt.legend()
plt.show()
```
